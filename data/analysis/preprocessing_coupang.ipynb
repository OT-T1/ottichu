```python
files = ['korean_drama', 'american_drama', 'asian_drama', 'britain_drama', 'documentary', 'female', 'romance', 'variety', 'web_drama']
names = ['한국 드라마', '미국 드라마', '아시아 드라마', '영국 드라마', '전체 다큐멘터리', '전체 여성 주인공', '전체 로맨스', '전체 예능', '전체 웹 드라마']
contents = [pd.DataFrame()*i for i in range(len(files))]

for idx, file in enumerate(files):
  file_path = f'{path}/coupang/prev/coupang_tv_{file}.json'
  with open(file_path, 'r') as json_file:
      json_data = json.load(json_file)

  i = 1
  while True:
    try : 
      num = f'{i}'
      contents[idx] = contents[idx].append(json_normalize(json_data[num]), ignore_index=True)
      i += 1
    except:
      break

  contents[idx] = contents[idx].drop(columns = ['age_rating'])
  contents[idx]['country'] = names[idx].split(' ')[0]
  contents[idx].rename(columns = {'image_link': 'poster_url', 'rating':'age_rating', "tags":"genre", 'as':'category'}, inplace=True)

#print(bri.head())
#print(bri.columns)
print(contents[8])
```

```python
tv = pd.DataFrame()
for content in contents:
  tv = tv.append(content, ignore_index=True)

print(tv.head())
print(tv.info)
```
